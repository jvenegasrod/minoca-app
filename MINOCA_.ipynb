## Domain Knowledge üìö

- **`age`**: This is the age of the patient. Age is a crucial factor in stroke prediction as the risk of stroke increases with age. According to the World Health Organization, the risk of stroke doubles every decade after the age of 55. üå°Ô∏èüë¥

- **`hypertension`**: This is a binary feature indicating whether the patient has hypertension (high blood pressure) or not. Hypertension is a significant risk factor for stroke as it can damage blood vessels, making them prone to blockage or rupture. üíâü©∫

- **`heart_disease`**: This binary feature indicates whether the patient has heart disease or not. Patients with heart diseases are at a higher risk of stroke as these conditions can lead to the formation of clots in the heart that can travel to the brain. ‚ù§Ô∏èüíî

- **`ever_married`**: This feature represents whether the patient is married or not. Although not a direct risk factor for stroke, marital status can be associated with lifestyle factors that influence stroke risk. For instance, married individuals might have different stress levels, physical activity patterns, or dietary habits compared to their unmarried counterparts. üíçüë´

- **`work_type`**: This categorical feature describes the type of occupation of the patient. Certain occupations might be associated with higher stress levels or sedentary behavior, which can influence stroke risk. üë∑‚Äç‚ôÄÔ∏èüë®‚Äç‚öïÔ∏èüë©‚Äçüíº

- **`Residence_type`**: This feature indicates whether the patient lives in a rural or urban area. The place of residence might be associated with stroke risk due to factors like access to healthcare, air quality, lifestyle habits, etc. üè°üåÜüå≥

- **`avg_glucose_level`**: This feature represents the average glucose level in the patient's blood. High blood glucose levels can damage blood vessels, leading to an increased risk of stroke. ü©∏üç¨

- **`bmi`**: This is the Body Mass Index of the patient, calculated as weight in kilograms divided by the square of height in meters. A high BMI indicates obesity, which is a significant risk factor for stroke as it can lead to or exacerbate conditions like hypertension, high blood glucose, and heart disease. üèãÔ∏è‚Äç‚ôÄÔ∏èüçî

- **`smoking_status`**: This categorical feature indicates whether the patient is a smoker, former smoker, or never smoked. Smoking can increase stroke risk as it can damage blood vessels, increase blood pressure, and reduce the amount of oxygen reaching the brain. üö¨üö≠

- **`gender`**: This feature represents the gender of the patient. Gender can influence stroke risk due to biological differences and gender-specific lifestyle patterns. ‚ôÄÔ∏è‚ôÇÔ∏è


### MICE:
Utilizamos MICE para imputar valores nulos
MICE (Multiple Imputation by Chained Equations) es una t√©cnica en aprendizaje autom√°tico para imputar valores faltantes en un conjunto de datos. En lugar de rellenar los valores nulos con un √∫nico valor (como la media o la mediana), MICE utiliza un enfoque m√°s complejo y detallado, imputando m√∫ltiples valores posibles para obtener una estimaci√≥n m√°s precisa. Aqu√≠ te explico c√≥mo funciona:

1.Imputaci√≥n inicial: Primero, se asigna un valor inicial para cada valor nulo, generalmente con la media, la mediana o alg√∫n valor aleatorio.

2.Modelo por cada columna: Luego, se imputa cada columna que tiene valores nulos usando un modelo que predice los valores faltantes de esa columna en funci√≥n de las otras columnas. Por ejemplo, si la columna A tiene valores nulos, MICE entrenar√° un modelo para predecir la columna A en funci√≥n de las dem√°s.

3.Iteraci√≥n en cadena: Este proceso se repite de forma iterativa para cada columna con valores faltantes, actualizando las imputaciones en cada paso. El algoritmo pasa varias veces sobre el conjunto de datos para mejorar la precisi√≥n de las imputaciones.

4.Imputaci√≥n m√∫ltiple: Finalmente, el proceso se repite varias veces (imputaci√≥n m√∫ltiple) para obtener diferentes conjuntos de datos imputados, que representan distintas posibilidades. Esto permite tener en cuenta la incertidumbre en la imputaci√≥n y generar un resultado m√°s robusto..

MICE es una opci√≥n s√≥lida porque puede imputar valores bas√°ndose en las relaciones entre m√∫ltiples variables. Esto es √∫til en datos cl√≠nicos, ya que las caracter√≠sticas (como edad, presi√≥n arterial, colesterol) suelen estar correlacionadas. ## Tratamiento muestra desbalanceada: SMOTE

¬øQu√© es SMOTE?
SMOTE es una t√©cnica que genera ejemplos sint√©ticos en los conjuntos de datos para la clase minoritaria (aquella con menos observaciones) en problemas de clasificaci√≥n. Su prop√≥sito es balancear el conjunto de datos para mejorar el rendimiento del modelo, ya que los modelos de aprendizaje supervisado suelen favorecer las clases con m√°s datos, lo que puede llevar a sesgos y menor precisi√≥n en la predicci√≥n de clases minoritarias.

¬øC√≥mo Funciona SMOTE?
Identificaci√≥n de Ejemplos de la Clase Minoritaria:

SMOTE comienza identificando las muestras de la clase minoritaria en el conjunto de datos. Supongamos que tienes un conjunto de datos con dos clases: "No infarto" y "Infarto", donde "Infarto" es la clase minoritaria.
Selecci√≥n de Vecinos M√°s Cercanos (KNN):

Para cada punto en la clase minoritaria, SMOTE selecciona algunos de sus vecinos m√°s cercanos. Generalmente se usa el algoritmo K-Nearest Neighbors (KNN) para encontrar, por ejemplo, los 5 vecinos m√°s cercanos de cada muestra en la clase minoritaria.
Interpolaci√≥n para Crear Ejemplos Sint√©ticos:

SMOTE crea nuevas observaciones interpolando entre la muestra seleccionada y uno de sus vecinos m√°s cercanos. Esto se hace al tomar un punto en la clase minoritaria y, en lugar de replicarlo directamente, genera un nuevo punto en una posici√≥n aleatoria entre √©l y un vecino.

Incorporaci√≥n de Ejemplos Sint√©ticos en el Conjunto de Datos:

Este proceso se repite para m√∫ltiples observaciones de la clase minoritaria, generando suficientes puntos sint√©ticos hasta igualar (o aproximarse) al n√∫mero de muestras en la clase mayoritaria.
Una vez generados, estos ejemplos sint√©ticos se agregan al conjunto de datos original, de modo que el modelo tenga datos balanceados para cada clase.
Ventajas de SMOTE
Mejora la precisi√≥n en la clase minoritaria: Al generar datos adicionales para la clase minoritaria, el modelo puede aprender mejor las caracter√≠sticas de esa clase, lo que reduce el sesgo hacia la clase mayoritaria.
Genera diversidad: A diferencia de la duplicaci√≥n directa, SMOTE produce ejemplos variados que ayudan a reducir el sobreajuste, permitiendo que el modelo generalice mejor.
F√°cil de implementar: Herramientas como imbalanced-learn en Python (imblearn) incluyen SMOTE, facilitando su integraci√≥n con scikit-learn.
Limitaciones de SMOTE
Posibles valores at√≠picos: Si existen valores at√≠picos en la clase minoritaria, SMOTE puede generar datos sint√©ticos en √°reas poco representativas, lo que podr√≠a afectar el rendimiento.
Relaciones lineales: SMOTE asume que las caracter√≠sticas pueden interpolarse de forma lineal, por lo que puede no ser ideal para datos que tengan relaciones complejas o no lineales.
Riesgo de sobreajuste: Generar demasiadas muestras sint√©ticas puede llevar al modelo a "memorizar" los datos, en lugar de generalizar. Esto es especialmente cr√≠tico cuando el dataset original es peque√±o y se generan muchos puntos sint√©ticos.

Estandarizamos los datos. Esto proporciona la media cero y la varianza unitaria de los datos:

-Alinea Escalas de las Variables: Algunas variables pueden tener diferentes escalas (por ejemplo, presi√≥n arterial frente a niveles de colesterol). Estandarizar evita que caracter√≠sticas con grandes escalas dominen a las de menor escala.

-Mejora el Rendimiento de Modelos Basados en Distancia: Algoritmos como k-Nearest Neighbors, Support Vector Machines, y m√©todos basados en redes neuronales, que dependen de la distancia entre datos, se benefician de la estandarizaci√≥n para medir de manera justa entre las diferentes caracter√≠sticas.

-Convergencia M√°s R√°pida en Algoritmos de Optimizaci√≥n: Modelos como redes neuronales y regresi√≥n log√≠stica, que usan optimizaci√≥n iterativa, suelen converger m√°s r√°pido cuando los datos est√°n estandarizados.

-Regularizaci√≥n y Penales en Modelos: Para modelos como regresi√≥n lineal o log√≠stica, que aplican regularizaci√≥n, estandarizar permite que el peso de cada variable est√© en la misma escala, ayudando al modelo a penalizar correctamente.

En resumen, estandarizar es importante para obtener modelos m√°s estables, precisos y eficientes, especialmente en algoritmos sensibles a la escala de los datos.

```python
import optuna
from sklearn.model_selection import cross_val_score

def objective(trial):
    # Sugerir hiperpar√°metros
    max_depth = trial.suggest_int("max_depth", 3, 20)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 20)
    
    # Modelo
    model = DecisionTreeClassifier(
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf
    )
    
    # Validaci√≥n cruzada
    scores = cross_val_score(model, X, y, cv=10, scoring="accuracy")
    mean_score = np.mean(scores)
    std_score = np.std(scores)
    
    # Penalizar desviaciones est√°ndar mayores a 0.05
    if std_score > 0.05:
        return mean_score - 0.1 * (std_score - 0.05)  # Penalizaci√≥n

    return mean_score

# Optimizaci√≥n
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

# Mejores par√°metros
print("Mejores par√°metros:", study.best_params)
print("Mejor puntuaci√≥n:", study.best_value)
```

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import cross_val_predict
from sklearn.preprocessing import label_binarize
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
# Par√°metros ajustados del modelo de Decision Tree
decision_tree = DecisionTreeClassifier(
    max_depth=20,               # Limita la profundidad m√°xima del √°rbol
    min_samples_split=4,      # Muestras m√≠nimas para dividir un nodo
    min_samples_leaf=6,        # Muestras m√≠nimas en una hoja
    max_features='sqrt',       # N√∫mero de caracter√≠sticas evaluadas en cada nodo
    class_weight='balanced',   # Para manejar el desbalance de clases
    random_state=42            # Para reproducibilidad
)

# Stratified K-Fold Cross-Validation
skf = StratifiedKFold(n_splits=5)

# Realiza la validaci√≥n cruzada y calcula la m√©trica de evaluaci√≥n (por ejemplo, exactitud)
scores = cross_val_score(decision_tree, X_p, y, cv=skf, scoring='accuracy')

print("Accuracy Scores in each fold:", scores)
print("Average Accuracy Score:", scores.mean())
mean_accuracy = np.mean(scores)
std_deviation = np.std(scores)

print(f"Precisi√≥n media: {mean_accuracy:.3f}")
print(f"Desviaci√≥n est√°ndar: {std_deviation:.3f}")
```

```python
# Inicializar listas para guardar las fpr, tpr y auc
fpr_list = []
tpr_list = []
roc_auc_list = []

# Binarizamos las etiquetas (en caso de que sea clasificaci√≥n binaria)
y_bin = label_binarize(y, classes=[0, 1])

# Realizar la validaci√≥n cruzada
for train_index, test_index in skf.split(X_p, y):
    X_train, X_test = X_p[train_index], X_p[test_index]
    y_train, y_test = y_bin[train_index], y_bin[test_index]

    # Entrenamos el modelo
    decision_tree.fit(X_train, y_train.ravel())  # .ravel() convierte en un vector unidimensional

    # Obtenemos las probabilidades de predicci√≥n (no las clases predichas)
    y_prob = decision_tree.predict_proba(X_test)[:, 1]  # Probabilidades para la clase positiva (1)

    # Calculamos la curva ROC y AUC para este fold
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    # Guardamos los valores de fpr, tpr y auc
    fpr_list.append(fpr)
    tpr_list.append(tpr)
    roc_auc_list.append(roc_auc)

# Promediamos las curvas ROC
mean_fpr = np.linspace(0, 1, 100)
mean_tpr = np.mean([np.interp(mean_fpr, fpr, tpr) for fpr, tpr in zip(fpr_list, tpr_list)], axis=0)

# Graficar la curva ROC promedio
plt.figure(figsize=(8, 6))
plt.plot(mean_fpr, mean_tpr, color='b', label=f'Mean ROC curve (AUC = 0.81)')

# Graficar las curvas ROC individuales para cada fold
for fpr, tpr, roc_auc in zip(fpr_list, tpr_list, roc_auc_list):
    plt.plot(fpr, tpr, lw=1, alpha=0.3, color='gray')

# Graficar la diagonal aleatoria (l√≠nea de referencia)
plt.plot([0, 1], [0, 1], color='black', linestyle='--')

# Etiquetas y t√≠tulo
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Stratified K-Fold Cross-Validation)')
plt.legend(loc='lower right')

# Mostrar la gr√°fica
plt.show()
```

### Construir la f√≥rmula de la puntuaci√≥n de riesgo

Para calcular la puntuaci√≥n de un paciente, sumamos los valores de sus variables multiplicados por los coeficientes normalizados:

Puntuaci√≥n: n=‚àë(valor¬†de¬†la¬†variable√ócoeficiente¬†normalizado)

üîπ ¬øC√≥mo interpretar la puntuaci√≥n?

Valores altos ‚Üí Mayor riesgo de MINOCA.

Valores bajos ‚Üí Menor riesgo.
